# Top 100 Best Movies Of All Time WebScraping Project

```python
from bs4 import BeautifulSoup
import requests
```

```python
path = "https://www.empireonline.com/movies/features/best-movies-2/"
```

```python
page = requests.get(path)
```

```python
page
```

```python
soup = BeautifulSoup(page.text , 'html')
```

```python
soup.prettify()
```

```python
movies_title = soup.find_all('h2')
```

```python
movies_title
```

```python
top_movies = [title.text.strip() for title in movies_title]
```

```python
top_movies
```

```python
import pandas as pd

# Filtering non-movie entries
ranked_movies = [line for line in top_movies if ')' in line and '(' in line and ')' in line.split('(')[0]]

data = []

for movie in ranked_movies:
    # Spliting to extract rank
    rank_part, rest = movie.split(')', 1)
    rank = int(rank_part.strip())

    # Find year from last brackets
    year_start = rest.rfind('(')
    year_end = rest.rfind(')')
    year = int(rest[year_start + 1:year_end])

    # Extracting movie name
    name = rest[:year_start].strip()

    data.append((rank, name, year))

# Creating DataFrame
df = pd.DataFrame(data, columns=['rank', 'name', 'year'])

df = df.sort_values(by='rank', ascending=False).reset_index(drop=True)

print(df.head())
```
